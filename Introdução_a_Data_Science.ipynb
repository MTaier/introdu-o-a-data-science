```python
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.original_language.unique()
```

```python
contagem_de_lingua = tmdb['original_language'].value_counts().to_frame().reset_index()
contagem_de_lingua.columns = ['original_language', 'total']
contagem_de_lingua.head()
```

```python
sns.barplot(data = contagem_de_lingua, x = 'original_language', y = 'total')
```

```python
sns.catplot(x= 'original_language', data = tmdb, kind = 'count')
```

```python
import matplotlib.pyplot as plt

plt.pie(contagem_de_lingua['total'], labels = contagem_de_lingua['original_language'])
```

```python
total_por_lingua = tmdb['original_language'].value_counts()
total_ingles = total_por_lingua['en']
total_geral = total_por_lingua.sum()
total_resto = total_geral - total_ingles
print(total_ingles, total_resto)
```

```python
dados = {
    'lingua': ['ingles', 'outros'],
    'total': [total_ingles, total_resto]
}
dados = pd.DataFrame(dados)

sns.barplot(dados, x = 'lingua', y = 'total')
```

```python
tmdb.query("original_language != 'en'").original_language.value_counts()
```

```python
filmes_sem_lingua_original_em_ingles = tmdb.query("original_language != 'en'")
```

```python
sns.catplot(filmes_sem_lingua_original_em_ingles,
            x = 'original_language',
            kind = 'count',
            aspect = 2,
            order = filmes_sem_lingua_original_em_ingles['original_language'].value_counts().index,
            palette="magma")
```
